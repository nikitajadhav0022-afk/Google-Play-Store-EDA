```python
##lets analyse this dataset by hping through each and every column and try to find some or the other information.
df.head(2)
```

```python
#chekcing if all value sin Reviews col are numeric or not
df['Reviews'].unique()
```

```python
#this means all records are basically numeric but they have string datatype.
#lets find out if all records are numeric or not.
df['Reviews'].str.isnumeric().sum()
```

```python
#above result shows 10840 as number of records present in Reviews column but in realtiy we have 10841 records in this dataset.
#which means there is 1 record in this column which is not numeric.
df[~df['Reviews'].str.isnumeric()]   #- this minus sign denotes not of df.. meaning its reverse. show values which are not numeric.
```

```python
#in above result we have 3.0M value in Reviews column which is not numeric.
#take a copy of your dataset -
df_copy=df.copy()
```

```python
#drop the faulty record from dataset copy
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
#convert Reviews datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('K','')  #for K will keep it blank

#trying to convert all million values into kunits i.e for 19million -> 19000000==19000k. replacing last 3 zeros with k that is 19 thousand k
```

```python
df_copy['Size'].isnull().sum()
```

```python
#but we still have 'Varies with device' value in Size..lets replace it with NaN value
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
#df_copy['Size']=df_copy['Size'].astype(float)  here it was giving error because we still had invalid values in dataset other than above M and K values
#so using pd.to_numeric because pd.to_numeric with errors='coerce' safely converts invalid entries to NaN, making it more robust for real-world datasets. 
df_copy['Size']=pd.to_numeric(df_copy['Size'],errors='coerce')
df_copy['Size']=df_copy['Size'].astype(float)   #now again trying to change datatype to float and it got chnaged
```

```python
df_copy.info()
```

```python
#we can replace this NaN value with something other, which we can do in feature eng part.

#cleaning wrt Installs
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
#in above 2 columns installs and price we can see special characters and some strings in dataset. replace special characters +,$ and , with blank.

chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
```

```python
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
#handling last updated col
df_copy['Last Updated'].unique()
```

```python
#convert last updated to date time datatype
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
```

```python
#creating 3 columns from last updated
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

## Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

## Insights
1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 14 Billion Installations GAME is the most popular Category in Google App store

## What are the Top 5 most installed Apps in Each popular Categories ??

```python
dfa = df_copy.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

## Insights
- Most popular game is Subway Surfers.
- Most popular communication app is Hangouts.
- Most popular productivity app is Google Drive.
- Most popular social app is Instagram.

## How many apps are there on Google Play Store which get 5 ratings??

```python
rating = df_copy.groupby(['Category','Installs', 'App'])['Rating'].sum().sort_values(ascending = False).reset_index()

toprating_apps = rating[rating.Rating == 5.0]
print("Number of 5 rated apps",toprating_apps.shape[0])
toprating_apps.head(1)
```

## Result
- There are 271 five rated apps on Google Play store
- Top most is 'CT Brain Interpretation' from 'Family' Category
